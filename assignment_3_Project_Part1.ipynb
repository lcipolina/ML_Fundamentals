### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Create two classea and calculate Gini, Entropy and Classification Error

```python
#Create two random variables (i.e. two classes)
pm1 = np.random.uniform(0.0, 1.0, 1000)
pm2 = 1 - pm1

#Calculate Gini, Entropy and Classification Error
E = 1 - np.maximum(pm1, pm2)                             #Error
G = np.add(np.multiply(pm1, pm2), np.multiply(pm1, pm2)) #Gini
D = np.add(-pm1 * np.log(pm1), -pm2 * np.log(pm2)).      #Entropy


#Save results on Dictionary to plot against pm1
E_dict = {}
G_dict = {}
D_dict = {}

for idx, pm in enumerate(pm1):
    E_dict[pm] = E[idx]
    G_dict[pm] = G[idx]
    D_dict[pm] = D[idx]
```

### Plot

```python
# Plot    

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

#Classification Error
lists = sorted(E_dict.items())
x, y = zip(*lists)

plt.plot(x, y, color='r', label='Classification Error')

#Gini
lists = sorted(G_dict.items())
x, y = zip(*lists)
plt.plot(x, y, color='g', label='Gini Index')

#Entropy
lists = sorted(D_dict.items())
x, y = zip(*lists)
plt.plot(x, y, color='y', label='Cross-Entropy')
ax.set_xlabel('pm1')
ax.set_ylabel('Gini index,Entropy,Errors')
ax.set_title('Error vs Probability')
plt.legend()
plt.grid()
plt.show()
```
